### Step 3: Identifying Duplicate Rows


**Task 1: Identify Duplicate Rows**
  1. Count the number of duplicate rows in the dataset.
  2. Display the first few duplicate rows to understand their structure.


```python
# Count full duplicate rows
duplicate_rows_count = df.duplicated().sum()
print(f"Total duplicate rows (entire row): {duplicate_rows_count}")

# Show duplicate rows
print(" "*40)
print("="*40)
duplicate_rows = df[df.duplicated()]

# print("\nFirst few duplicate rows:")

duplicate_rows = df[df['ResponseId'].duplicated(keep=False)]
duplicate_rows = duplicate_rows.sort_values(by='ResponseId')
duplicate_rows.head()
```

### Step 4: Removing Duplicate Rows


**Task 2: Remove Duplicates**
   1. Remove duplicate rows from the dataset using the drop_duplicates() function.
2. Verify the removal by counting the number of duplicate rows after removal .


```python
## 1. Write your code here

print(f"Total rows before removing duplicates: {len(df)}")

# Remove duplicates based on ResponseId
df_cleaned = df.drop_duplicates(subset='ResponseId')
# Remove duplicates based on ResponseId
df = df.drop_duplicates(subset='ResponseId')

print("="*45)
print(f"Total rows after removing duplicates: {len(df_cleaned)}")

# Count full duplicate rows
print("="*45)

duplicate_rows_count = df.duplicated().sum()
print(f"Total duplicate rows (entire row): {duplicate_rows_count}")
```

### Step 5: Handling Missing Values


**Task 3: Identify and Handle Missing Values**
   1. Identify missing values for all columns in the dataset.
   2. Choose a column with significant missing values (e.g., EdLevel) and impute with the most frequent value.


```python
## 1. Write your code here

# Count missing values per column
missing_values = df.isnull().sum()
missing_values_sorted = missing_values.sort_values(ascending=False)
print("Missing values per column:")
print(missing_values_sorted)

## 2. Write your code here
print("="*60)
print("")

## check how many are missing
print(f"EdLevel is missing: {df['EdLevel'].isnull().sum()}")

## Find the most frequent value (mode)
print("="*60)
print("")

most_frequent = df['EdLevel'].mode()[0]

print(f"Most frequent value in EdLevel: {most_frequent}")

## Impute missing values
print("="*60)
print("")

df['EdLevel'] = df['EdLevel'].fillna(most_frequent)

## verify 
print("="*60)
print("")

print("Missing values after imputation:")
print(df['EdLevel'].isnull().sum())
```

### Step 6: Normalizing Compensation Data


**Task 4: Normalize Compensation Data Using ConvertedCompYearly**
   1. Use the ConvertedCompYearly column for compensation analysis as the normalized annual compensation is already provided.
   2. Check for missing values in ConvertedCompYearly and handle them if necessary.


```python
## Write your code here

## Check missing vaprint("Missing values in ConvertedCompYearly:")
print("Missing values in ConvertedCompYearly:")
print(df['ConvertedCompYearly'].isnull().sum())

## Inspect the column - help understands: distribution, outliers, whether mean/median makes sense

print("="*60)
print("")

print(df['ConvertedCompYearly'].describe())

## Handle missing values
##Since this is numerical data, best practice = median (robust to outliers)
print("="*60)
print("")
median_comp = df['ConvertedCompYearly'].median()

print(f"Median compensation: {median_comp}")

## Impute missing values
print("="*60)
print("")

df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_comp)

# verify
print("Missing values after imputation:")
print(df['ConvertedCompYearly'].isnull().sum())
```
